```python
#2. From the second text file (file2.txt), write Python code and use MapReduct to count 
#how many times non-English words (names, places, spells etc.) were used. List 
#those words and how many times each was repeated.


def load_eng_words(file_path):
#dob : 04/04/2002
    #Loading  English words from a file and store them .

   
        
    eng_words = set()
    with open(file_path, 'r', encoding='utf-8') as engfile:
        for line in engfile:
            word = line.strip().lower()
            eng_words.add(word)
    return eng_words
 #Returns: Set containing English words in lowercase. 

def count_non_eng_words(file_path, eng_words):
    
    #Count non-English words in a file.

   
    non_eng = {}
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            words = line.strip().split()
            for word in words:
                word = word.lower()
                if word not in eng_words:
                    non_eng[word] = non_eng.get(word, 0) + 1
    return non_eng

if __name__ == "__main__":
    # Load English words from a file
    eng_words = load_eng_words("usa.txt")

    # Counting non-English words in file2.txt
    file_path = "file2.txt"
    non_eng_word_counts = count_non_eng_words(file_path, eng_words)

    # Printing non-English words
    for word, count in sorted(non_eng_word_counts.items()):
        print(f"{word}: {count}")
```
